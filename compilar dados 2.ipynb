```python
import pandas as pd
```

```python
# Carrega o arquivo Excel em um objeto Pandas
xl = pd.ExcelFile('Fluxo de Pagamento Emprestimos I.xlsx')

# Cria um objeto Pandas DataFrame vazio para armazenar os dados compilados
df = pd.DataFrame()

# Itera através de cada aba no arquivo Excel e adicione as informações à DataFrame
for sheet_name in xl.sheet_names:
    sheet_data = xl.parse(sheet_name)
    sheet_data['Nome_Aba'] = sheet_name  # Adiciona uma coluna com o nome da aba
    df = pd.concat([df, sheet_data], ignore_index=True)

# Salva o DataFrame compilado em um novo arquivo Excel
df.to_excel('Fluxo de Pagamento Emprestimos_compilado2.xlsx', index=False)
```

```python
# Carrega o arquivo Excel compilado em um objeto Pandas
compilada = pd.read_excel('Fluxo de Pagamento Emprestimos_compilado2.xlsx')
for index, row in compilada.iterrows():
    if pd.isna(row['Nome do Componente']):
        aba = row['Nome_Aba']
        componente = compilada.loc[(compilada['Nome_Aba'] == aba) & (~pd.isna(compilada['Nome do Componente']))]['Nome do Componente'].values
        if len(componente) > 0:
            compilada.at[index, 'Nome do Componente'] = componente[0]


compilada.head(25)
```

```python
# Filtra as linhas que possuem "PRI" na coluna "Tipo"
compilada = compilada.loc[compilada['Tipo'] == 'PRI']

# Seleciona apenas as colunas "Data Pagamento", "Calculado", "Nome do Componente" e "Nome_Aba"
compilada_final = compilada[['Data Pagamento', 'Tipo' ,'Calculado (convertido)', 'Nome do Componente', 'Nome_Aba']]


#Não há necessidade da informação do financiamento TJLP, sendo melhor exlcuir
compilada_final = compilada_final[~compilada_final['Nome do Componente'].str.contains('TJLP')]

#Separar qual é a taxa de juros atrelado apenas ao IPCA
compilada_final[['IPCA', 'Valor %']] = compilada_final['Nome do Componente'].str.split('+', expand=True)

compilada_final['Valor %'] = compilada_final['Valor %'].str.replace('+', '').str.strip().str.replace(' ', '')



display(compilada_final)
```

```python
compilada_final.to_excel('compilado2.xlsx', index=False)
```
